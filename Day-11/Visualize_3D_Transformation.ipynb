```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

```

```python
# Create 3D points
points = np.array([
    [0, 0, 0],
    [1, 0, 0],  # x-axis
    [0, 1, 0],  # y-axis
    [0, 0, 1],  # z-axis
])
```

```python
# Apply a rotation around Z axis
theta = np.radians(45)
rotation_matrix = np.array([
    [np.cos(theta), -np.sin(theta), 0],
    [np.sin(theta), np.cos(theta),  0],
    [0,             0,              1]
])
rotated_points = points @ rotation_matrix.T
```

```python
# Plot original and rotated points
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Original frame
ax.quiver(0, 0, 0, 1, 0, 0, color='r', label='X')
ax.quiver(0, 0, 0, 0, 1, 0, color='g', label='Y')
ax.quiver(0, 0, 0, 0, 0, 1, color='b', label='Z')

# Rotated frame
ax.quiver(0, 0, 0, *rotation_matrix[:,0], color='r', linestyle='dashed')
ax.quiver(0, 0, 0, *rotation_matrix[:,1], color='g', linestyle='dashed')
ax.quiver(0, 0, 0, *rotation_matrix[:,2], color='b', linestyle='dashed')

ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([0, 1])
ax.set_title('Original vs Rotated 3D Axes')
plt.legend()
plt.show()
```
